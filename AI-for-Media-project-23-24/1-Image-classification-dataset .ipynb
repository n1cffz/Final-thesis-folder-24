### Understanding ratio of images for each emotional class
The origional ratio submitted for AI-for-Media-project 2024 consisted of the following:
- Neutral-face images: 45
- Happy-face images: 285
- Sad-face images: 219
- Angry-face images: 156

```python
# Import more sad, neutral and angry faces now using Google search to populate each class with at least over 1000 images each, ensuring that only human faces were present( 'sad human faces')

# Define the search terms and corresponding subfolders in my-classification-dataset.
# Key words such as 'potrait, 'photo' and variations of searches were added as a search_term.
search_terms = {
    #"happy-face": "happy person face",
    #"sad-face":  "lonely person face" #"depressed person face" #"melancholic expression" #sad elderly person" #"sorrowful face" #"crying person" #"sad photo man",
    "angry-face": "enraged expression" #angry potrait",
    #"neutral-face": "neutral person"
    
}

# Base directory for saving images
base_directory = "/Users/nixi/Desktop/Final-thesis-folder-24/AI-for-Media-project-23-24/my-classification-dataset"

for folder_name, search_term in search_terms.items(): # Code adapted from https://stackoverflow.com/questions/273192/how-do-i-create-a-directory-and-any-missing-parent-directories
    folder_path = os.path.join(base_directory, folder_name)
    os.makedirs(folder_path, exist_ok=True)

   # Create a GoogleImageCrawler instance using code adapted from : https://snyk.io/advisor/python/icrawler/functions/icrawler.builtin.GoogleImageCrawler 
    crawler = GoogleImageCrawler(storage={'root_dir': folder_path})
    crawler.crawl(keyword=search_term, max_num=1000)  
```

```python
#I ran into an error where iCrawler had placed a limit on the number of searches,and a majority of images were skipped under the pretence of appearing as a duplicate in my dataset already saved. 
# sets up a user agent for GoogleImageCrawler, with specified threading configurations to efficiently download images into designated directories.
# The script ensures that images are filtered by size and can overwrite existing files if needed. Adjust the search terms and parameters to refine the image collection process. Set up the crawler with a custom user-agent and delay
# Code adapted from ; https://garudabyte.com/how-to-crawl-images-from-google/ 
crawler = GoogleImageCrawler(
    storage={'root_dir': folder_path},
    downloader_threads=4,  # Number of threads for downloading
    parser_threads=2,      # Number of threads for parsing
    feeder_threads=1       # Number of threads for feeding URLs
)

# Crawl with the specified settings
crawler.crawl(
    keyword=search_term,
    max_num=1000,  # Adjust as needed
    min_size=(200, 200),  # Minimum size to filter out thumbnails
    max_size=None,
    file_idx_offset='auto',
    overwrite=True,  # Overwrite existing files
    # delay=2.0  # Delay between requests in seconds (if supported)
)
```

After introducing new methods to populate each emotional class folder emotional class ratio are as follows:

```python
# Find out the total images in each emotional class within my current directory

emotional_classes = ['neutral-face', 'happy-face', 'sad-face', 'angry-face', ]

dirrectory = "/Users/nixi/Desktop/Final-thesis-folder-24/AI-for-Media-project-23-24/my-classification-dataset"

#Instantiate for counts
class_counts = {}

for emotion in emotional_classes:
    class_dir = os.path.join(dirrectory, emotion)
    # Count the number of image files in the class directory
    class_counts[emotion] = len([file for file in os.listdir(class_dir) if file.endswith(('jpg', 'jpeg', 'png', 'gif', 'bmp'))])


# Print out the value of each folder
for emotion_class, count in class_counts.items():
    print(f"Emotional class: {emotion_class}, Total images: {count}")
    class_counts[emotion] = len([file for file in os.listdir(class_dir) if file.endswith(('jpg', 'jpeg', 'png', 'gif', 'bmp'))])
```

```python
# Plotting the class balance 
plt.figure(figsize=(10, 6))
plt.bar(class_counts.keys(), class_counts.values(), color=['blue', 'orange', 'green', 'red'])
plt.title('Total Images in Each Emotional Class')
plt.xlabel('Emotional Classes')
plt.ylabel('Number of Images')
plt.xticks(rotation=45)
plt.grid(axis='y')


plt.tight_layout()
plt.show()
```
